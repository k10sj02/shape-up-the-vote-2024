```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
zipcodes = pd.read_csv("zipcodes.csv", encoding="utf-8", encoding_errors="ignore")
zipcodes.head()
```

```python
# Split the 'common_city_list' column into multiple columns
split_columns = zipcodes['common_city_list'].str.split(',', expand=True)

# Retain only the first column (0) after the split
zipcodes['common_city_list'] = split_columns[0]

# Strip leading and trailing whitespace from the 'common_city_list' column
zipcodes['common_city_list'] = zipcodes['common_city_list'].str.strip()

# Remove square brackets and quotes from the 'common_city_list' column
zipcodes['common_city_list'] = zipcodes['common_city_list'].str.replace('[', '', regex=False)
zipcodes['common_city_list'] = zipcodes['common_city_list'].str.replace(']', '', regex=False)
zipcodes['common_city_list'] = zipcodes['common_city_list'].str.replace("'", '', regex=False)

#Remove county string from county column
zipcodes['county'] = zipcodes['county'].str.replace("County", '', regex=False)

# Display the first few rows of the modified DataFrame
zipcodes.head()
```

```python
# Print a specific number of items from the 'population_by_year' column
num_items_to_print = 3  # Change this to the desired number of items
for item in zipcodes['population_by_year'].head(num_items_to_print):
    if isinstance(item, dict):
        for year, population in item.items():
            print(f"Year: {year}, Population: {population}")
    else:
        print("Invalid item:", item)
```
