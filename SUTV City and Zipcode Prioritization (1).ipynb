```python
# Import Libraries and Set Options

import pandas as pd
import numpy as np
import json
# Import the warnings module to handle warnings
import warnings
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

# Load Data

zipcodes = pd.read_csv("zipcodes.csv", encoding="utf-8", encoding_errors="ignore")
# zipcodes.head()
```

```python
import warnings
import pandas as pd

# Set the filter to ignore all warnings
warnings.simplefilter(action='ignore', category=Warning)

# Define a function to extract a specific value from a dictionary column
def get_key_val(col: str, key_val: str) -> str:
    # Check if the column value is null
    if pd.isnull(col): 
        return ''  # Return an empty string if the column value is null
    # Evaluate the string representation of the dictionary to get the actual dictionary
    val_dict = eval(col)[0]['values']
    # Extract the value associated with the specified key
    val_list = [item['y'] for item in val_dict if item['x'] == key_val]
    # if not val_list:
    #     print(f"No value found for key '{key_val}' in column data: {val_dict}")  # Debug statement
    return val_list[0] if val_list else ''  # Return the first item in the list or an empty string if the list is empty

# Define a function to explode a dictionary column into multiple columns
def explode_col(df: pd.DataFrame, col: str) -> pd.DataFrame:
    """Returns a dataframe with new columns exploded from the original"""
    # Ensure the column to be exploded exists in the dataframe
    assert col in df.columns.tolist()
    # Evaluate the string representation of the dictionary to get the actual dictionary
    val_dict = eval(df[col].iloc[0])[0]['values']
    # Extract all the keys from the dictionary
    val_keys = [item['x'] for item in val_dict]
    # print(f"Exploding column '{col}' with keys: {val_keys}")  # Debug statement
    # For each key, create a new column with the corresponding values
    for vk in val_keys:
        new_col_name = col + '_' + str(vk)
        # print(f"Creating new column: {new_col_name}")  # Debug statement
        df[new_col_name] = df[col].apply(lambda col: get_key_val(col, vk))
    return df  # Return the modified dataframe

# Define a function to get the total of exploded columns
def get_exploded_col_total(df: pd.DataFrame, col: str) -> pd.DataFrame:
    # Get all columns that start with the specified prefix
    exploded_cols = [c for c in df.columns.tolist() if c.startswith(f'{col}_')]
    # print(f"Calculating total for columns: {exploded_cols}")  # Debug statement
    # Calculate the sum of the specified columns and create a new column for the total
    df[f'{col}_total'] = df.loc[:, exploded_cols].sum(axis=1)
    return df  # Return the modified dataframe

# Select specific columns from the zipcodes dataframe
exp_df = zipcodes[['zipcode', 
                   'state', 
                   'median_household_income', 
                   'population_by_race',
                   'population_by_age', 
                   'population_by_year'
                  ]]

# Print the columns to verify
# print("Columns before exploding:", exp_df.columns.tolist())  # Added this line to print column names

# Explode the 'population_by_race' column into multiple columns
exp_df = explode_col(exp_df, 'population_by_race')
# Calculate the total of the exploded 'population_by_race' columns
exp_df = get_exploded_col_total(exp_df, 'population_by_race')
# Delete the original 'population_by_race' column from the dataframe
del exp_df['population_by_race']

# Explode the 'population_by_age' column into multiple columns
exp_df = explode_col(exp_df, 'population_by_age')
# Calculate the total of the exploded 'population_by_age' columns
exp_df = get_exploded_col_total(exp_df, 'population_by_age')
# Delete the original 'population_by_age' column from the dataframe
del exp_df['population_by_age']

# Print the columns again to verify the changes
# print("Columns after exploding 'population_by_race' and 'population_by_age':", exp_df.columns.tolist())  # Added this line to print column names again

# Explode the 'population_by_year' column into multiple columns
exp_df = explode_col(exp_df, 'population_by_year')
# Calculate the total of the exploded 'population_by_year' columns
exp_df = get_exploded_col_total(exp_df, 'population_by_year')
# Delete the original 'population_by_year' column from the dataframe
del exp_df['population_by_year']

# Print the final columns to verify the changes
# print("Final columns after exploding 'population_by_year':", exp_df.columns.tolist())  # Added this line to print final column names
```

```python
def calc_nonwhite_percentage(row) -> float: 
    if row['population_by_race_White'] == '' or row['population_by_race_total'] == '':
        return None
    return 1 - int(row['population_by_race_White']) / int(row['population_by_race_total'])

exp_df['population_by_race_nonwhite_percentage'] = exp_df.apply(lambda row: calc_nonwhite_percentage(row), axis=1)

# exp_df['high_priority_zip'] = exp_df.apply(lambda row: row['population_by_race_nonwhite_percentage'] > 0.4 and row['median_household_income'] < 50000, axis=1)
exp_df['high_priority_zip'] = exp_df.apply(lambda row: row['population_by_race_nonwhite_percentage'] > 0.33, axis=1)
```
